```python
core_weather["month_max"] = core_weather["TEMP MAX DEG"].rolling(30).mean()

core_weather["month_day_max"] = core_weather["month_max"] / core_weather["TEMP MAX DEG"]

core_weather["max_min"] = core_weather["TEMP MAX DEG"] / core_weather["TEMP MIN DEG"]
```

```python
core_weather = core_weather.iloc[30:,:].copy()
```

```python
def create_predictions(predictors, core_weather, reg):
    train = core_weather.loc[:"12/31/2018"]
    test = core_weather.loc["1/1/2019":]

    reg.fit(train[predictors], train["target"])
    predictions = reg.predict(test[predictors])

    error = mean_squared_error(test["target"], predictions)
    
    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ["actual", "predictions"]
    return error, combined
```

```python
from sklearn.metrics import mean_squared_error
```

```python
predictors = ["RAIN SUM MM", "TEMP MAX DEG", "TEMP MIN DEG", "month_day_max", "max_min"]

error, combined = create_predictions(predictors, core_weather, reg)
error
```

```python
combined.plot()
```

```python
core_weather["monthly_avg"] = core_weather["TEMP MAX DEG"].groupby(core_weather.index.month).apply(lambda x: x.expanding(1).mean())
core_weather["day_of_year_avg"] = core_weather["TEMP MAX DEG"].groupby(core_weather.index.day_of_year).apply(lambda x: x.expanding(1).mean())
```

```python
error, combined = create_predictions(predictors + ["monthly_avg", "day_of_year_avg"], core_weather, reg)
error
```
